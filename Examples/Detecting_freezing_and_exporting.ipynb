# Detecting freeznig behavior and exporting for visualization and data analysis

## Import necessary packages

```python
# Import essential modules
import os
import pandas as pd
import FrozenPy.functions as fp
```

## Load data

```python
# read .out files from data directory
data_dir = '/Users/home/Documents/Jupyter/my-modules/FrozenPy/Examples/data'
fp.read_out(data_dir)

# change current directory to data directory
os.chdir(data_dir)

# Load Conditioning data
Cond_S1 = fp.read_rawcsv('Example_Conditioning_Sq1.raw.csv')
Cond_S2 = fp.read_rawcsv('Example_Conditioning_Sq2.raw.csv')
Cond_S3 = fp.read_rawcsv('Example_Conditioning_Sq3.raw.csv')

# Load Extinction data
Ext_S1 = fp.read_rawcsv('Example_Extinction_Sq1.raw.csv')
Ext_S2 = fp.read_rawcsv('Example_Extinction_Sq2.raw.csv')
Ext_S3 = fp.read_rawcsv('Example_Extinction_Sq3.raw.csv')

# Load Retrieval data
Ret_S1 = fp.read_rawcsv('Example_Retrieval_Sq1.raw.csv')
Ret_S2 = fp.read_rawcsv('Example_Retrieval_Sq2.raw.csv')
Ret_S3 = fp.read_rawcsv('Example_Retrieval_Sq3.raw.csv')
```

## Detect freezing

### Conditioning

Detect freezing for conditioning data

```python
# Freezing to Threshold
Cond_Sq1_df = fp.detect_freezing(Cond_S1)
Cond_Sq2_df = fp.detect_freezing(Cond_S2)
Cond_Sq3_df = fp.detect_freezing(Cond_S3)

# Concatenate all squad dataframes
Cond_df = pd.concat([Cond_Sq1_df, Cond_Sq2_df, Cond_Sq3_df], axis=1)
Cond_df.head()
```

### Extinction

Detect freezing for extinction data

```python
# Freezing to Threshold
Ext_Sq1_df = fp.detect_freezing(Ext_S1)
Ext_Sq2_df = fp.detect_freezing(Ext_S2)
Ext_Sq3_df = fp.detect_freezing(Ext_S3)

# Concatenate all squad dataframes
Ext_df = pd.concat([Ext_Sq1_df, Ext_Sq2_df, Ext_Sq3_df], axis=1)
Ext_df.head()
```

### Retrieval

Detect freezing for retrieval data

```python
# Freezing to Threshold
Ret_Sq1_df = fp.detect_freezing(Ret_S1)
Ret_Sq2_df = fp.detect_freezing(Ret_S2)
Ret_Sq3_df = fp.detect_freezing(Ret_S3)

# Concatenate all squad dataframes
Ret_df = pd.concat([Ret_Sq1_df, Ret_Sq2_df, Ret_Sq3_df], axis=1)
Ret_df.head()
```

## Slice data 

Now freezing episodes have been detected throughout the entire session, we need to extract the timepoints that we care out. Namely, the 3 minute baseline period and CS trials (both the CS and ISI) averaged across time.

Because the CS trials is a repeated measure, and the baseline is not, these data points are stored in separate variables; BL_df, and Trials_df, respectively. This makes handling pairing with metadata much easier.

### Conditioning

Slice conditioning data

```python
Cond_BL_df, Cond_Trials_df = fp.get_averagedslices(df=Cond_df, # data to analyze
                                         BL=180,               # 180s baseline
                                         CS=10,                # 10s CS
                                         US=2,                 # 2s US
                                         Trials=5,             # 5 trials
                                         ISI=58,               # 58s ISI
                                         fs=5,                 # 5 Hz sampling frequency
                                         Behav='Freezing')     # get freezing data
display(Cond_BL_df.head())
Cond_Trials_df.head()
```

### Extinction

Slice extinction data

```python
Ext_BL_df, Ext_Trials_df = fp.get_averagedslices(df=Ext_df, # data to analyze
                                         BL=180,            # 180s baseline
                                         CS=10,             # 10s CS
                                         US=0,              # 0s US
                                         Trials=45,         # 45 trials
                                         ISI=30,            # 30s ISI
                                         fs=5,              # 5 Hz sampling frequency
                                         Behav='Freezing')  # get freezing data
display(Ext_BL_df.head())
Ext_Trials_df.head()
```

### Retrieval

Slice retrieval data

```python
Ret_BL_df, Ret_Trials_df = fp.get_averagedslices(df=Ret_df, # data to analyze
                                         BL=180,            # 180s baseline
                                         CS=10,             # 10s CS
                                         US=0,              # 0s US
                                         Trials=45,         # 45 trials
                                         ISI=30,            # 30s ISI
                                         fs=5,              # 5 Hz sampling frequency
                                         Behav='Freezing')  # get freezing data
display(Ret_BL_df.head())
Ret_Trials_df.head()
```

## Merge data with metadata

To allow for reliable visualization and statistical testing, we should merge the freezing data with metadata regarding the subjects. This is done by creating a simple Excel sheet with the desired metadata, importing that data into a pandas dataframe, then merge your data and metadata dataframes. 

```python
# load metadata
metadata_dir = '/Users/home/Documents/Jupyter/my-modules/FrozenPy/Examples/data'
Metadata_df = fp.read_metadata(metadata_dir, sep=',')
```

### Conditioning

Merge conditioning data

```python
Cond_BL_md_df = pd.merge(Metadata_df, Cond_BL_df, on='ID', copy='True')
Cond_Trials_md_df = pd.merge(Metadata_df, Cond_Trials_df, on='ID', copy='True')

display(Cond_Trials_md_df.head())
```

### Extinction

Merge extinction data

```python
Ext_BL_md_df = pd.merge(Metadata_df, Ext_BL_df, on='ID', copy='True')
Ext_Trials_md_df = pd.merge(Metadata_df, Ext_Trials_df, on='ID', copy='True')

display(Ext_Trials_md_df.head())
```

### Retrieval

Merge retrieval data

```python
Ret_BL_md_df = pd.merge(Metadata_df, Ret_BL_df, on='ID', copy='True')
Ret_Trials_md_df = pd.merge(Metadata_df, Ret_Trials_df, on='ID', copy='True')

Ret_Trials_md_df.head()
```

## Convert to wide format and export

Before exporting it may be useful to convert from long to wide format. This is easier to work with in Excel and is the prefered format for some statistical softwares (Statview).

### Conditioning 

Converting and exporting conditioning data

```python
# Pivot long to wide
Cond_Trials_wide = Cond_Trials_md_df.pivot_table(index=['ID', 'Sex', 'Group'],
                                       columns='Trial',
                                       values=['CS', 'US', 'ISI'])
# display wide
display(Cond_Trials_wide.head())

# export to csv
Cond_Trials_md_df.to_csv('Conditioning_Trials_Long.csv')
Cond_BL_md_df.to_csv('Conditioning_BL_Long.csv')
Cond_Trials_wide.to_csv('Conditioning_Freezing_Wide.csv')
```

### Extinction

Converting and exporting extinction data

```python
# Pivot long to wide
Ext_Trials_wide = Ext_Trials_md_df.pivot_table(index=['ID', 'Sex', 'Group'],
                                       columns='Trial',
                                       values=['CS', 'US', 'ISI'])
# display wide
display(Ext_Trials_wide.head())

# export to csv
Ext_Trials_md_df.to_csv('Extinction_Trials_Long.csv')
Ext_BL_md_df.to_csv('Extinction_BL_Long.csv')
Ext_Trials_wide.to_csv('Extinction_Freezing_Wide.csv')
```

### Retrieval

Converting and exporting Retrieval data

```python
# Pivot long to wide
Ret_Trials_wide = Ret_Trials_md_df.pivot_table(index=['ID', 'Sex', 'Group'],
                                       columns='Trial',
                                       values=['CS', 'US', 'ISI'])
# display wide
display(Ret_Trials_wide.head())

# export to csv
Ret_Trials_md_df.to_csv('Retrieval_Trials_Long.csv')
Ret_BL_md_df.to_csv('Retrieval_BL_Long.csv')
Ret_Trials_wide.to_csv('Retrieval_Freezing_Wide.csv')
```
