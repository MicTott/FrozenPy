## Read .raw.csv and detect freezing

```python
# read .raw.csv
os.chdir(CS_data_dir)
CS_data_raw = fp.read_rawcsv('20200323 - CRH IED - Cond - Sq 1.raw.csv')

# detect freezing
CS_data_frz = fp.detect_freezing(CS_data_raw)
```

## Slice data

```python
# slice data
CS_bl, CS_trials = fp.get_averagedslices(df=CS_data_frz,
                                                           BL=180,
                                                           CS=10,
                                                           US=2,
                                                           Trials=5,
                                                           ISI=58,
                                                           fs=5,
                                                           Behav='Freezing')
CS_trials.head()
```

```python
# pivot long to wide
CS_trials_wide = CS_trials.pivot_table(index='ID',
                                         columns=['Trial'],
                                         values=['CS', 'US', 'ISI'])
CS_trials_wide
```
